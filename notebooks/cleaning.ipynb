# Dealing with missing data

```python
data.isna().sum()
```

#### I drop the column "confidence_in_national_government" because the values in this column are all NAN.

```python
data.drop(['confidence_in_national_government'], axis=1, inplace=True)
```

```python
data
```

```python
data.isna().sum()
```

#### I fill out the missing values of these columns with the mean because the difference between the average value and the actual value is small. 

```python
data['log_gdp_per_capita'].fillna(data['log_gdp_per_capita'].mean(), inplace=True)
data['healthy_life_expectancy_at_birth'].fillna(data['healthy_life_expectancy_at_birth'].mean(), inplace=True)
data['freedom_to_make_life_choices'].fillna(data['freedom_to_make_life_choices'].mean(), inplace=True)
data['generosity'].fillna(data['generosity'].mean(), inplace=True)
data['perceptions_of_corruption'].fillna(data['perceptions_of_corruption'].mean(), inplace=True)
```

```python
data.isna().sum()
```

#### I drop the missing 3 values of the column 'regional_indicator' , because this data is neither important nor the number of missing values is very small

```python
data.dropna(inplace = True)
```
